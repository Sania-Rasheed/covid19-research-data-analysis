## To Create Token Table

```python
df_1 = df.groupby(['paper_id', 'section_num'])['text'].apply(lambda x: x.str.split(r'\n\n+', expand=True).stack()).to_frame()
df_1
```

```python
df_1 = df_1.reset_index()
```

```python
df_1 = df_1.rename(columns={'level_2':'extra','level_3':'para_num'})
df_1
```

```python
df_1 = df_1.drop('extra', axis=1)
```

```python
df_1
```

```python
df_2 = df_1.groupby(['paper_id', 'section_num', 'para_num'])['text'].apply(lambda x: x.str.split(r'[.?!;"]', expand=True).stack()).to_frame()
df_2
```

```python
df_2 = df_2.reset_index()
df_2
```

```python
df_2 = df_2.rename(columns={'level_3':'extra','level_4':'sent_num'})
df_2
```

```python
df_2 = df_2.drop('extra', axis=1)
df_2
```

```python
df_2['tsent_num'] = range(len(df_2))
df_2
```

```python
df_2.drop(df_2.tail(1).index,inplace=True)
```

```python
df_2
```

```python
def preprocess(text):
    return [word for word in word_tokenize(text) if word.isalnum()]
```

```python
df_3 = df_2['text'].apply(preprocess).apply(pd.Series).stack().to_frame().reset_index()
```

```python
df_3 = df_3.rename(columns={'level_0':'tsent_num','level_1':'token_num', 0 : 'token_str'})
df_3
```

```python
TOKEN = df_2.merge(df_3, left_on='tsent_num', right_on='tsent_num')
```

```python
TOKEN = TOKEN.drop(['text', 'tsent_num'], axis = 1)
```

```python
TOKEN
```

```python
token_list = list(TOKEN.token_str)
token_list
```

```python
from nltk import pos_tag
```

```python
TOKEN['pos_tuple'] = pos_tag(token_list)
```

```python
TOKEN['pos'] = [x[1] for x in TOKEN.pos_tuple]
```

```python
TOKEN['term_str'] = TOKEN['token_str'].str.lower().str.replace('[\W_]', '')
```

```python
TOKEN
```

```python
TOKEN['term_str'].replace('', np.nan, inplace=True)
```

```python
TOKEN = TOKEN.dropna(subset=['term_str'])
```

```python
TOKEN
```

# VOCAB table

```python
VOCAB = TOKEN.term_str.value_counts().to_frame()\
    .rename(columns={'index':'term_str', 'term_str':'n'})\
    .sort_index().reset_index().rename(columns={'index':'term_str'})
VOCAB.index.name = 'term_id'
```

```python
VOCAB['num'] = VOCAB.term_str.str.match("\d+").astype('int')
VOCAB
```

```python
sw = pd.DataFrame(nltk.corpus.stopwords.words('english'), columns=['term_str'])
sw = sw.reset_index().set_index('term_str')
sw.columns = ['dummy']
sw.dummy = 1
sw.head
```

```python
VOCAB['stop_words'] = VOCAB.term_str.map(sw.dummy)
VOCAB['stop_words'] = VOCAB['stop_words'].fillna(0).astype('int')
```

```python
VOCAB[VOCAB.stop_words == 0].sample(10)
```

```python
VOCAB[VOCAB.stop_words == 1].sample(10)
```

```python
from nltk.stem.porter import PorterStemmer
stemmer1 = PorterStemmer()
VOCAB['stem_porter'] = VOCAB.term_str.apply(stemmer1.stem)

from nltk.stem.snowball import SnowballStemmer
stemmer2 = SnowballStemmer("english")
VOCAB['stem_snowball'] = VOCAB.term_str.apply(stemmer2.stem)

from nltk.stem.lancaster import LancasterStemmer
stemmer3 = LancasterStemmer()
VOCAB['stem_lancaster'] = VOCAB.term_str.apply(stemmer3.stem)
```

```python
VOCAB.to_csv('../dataset/processed-files/VOCAB.csv', index=False)
```
